### **Changing the categorical variable**

```python
df['Gender'].replace(['Male', 'Female'],
                        [0, 1], inplace=True)
df['Attrition'].replace(['No', 'Yes'],
                        [0, 1], inplace=True)
df['OverTime'].replace(['No', 'Yes'],
                        [0, 1], inplace=True)
```

### **One Hot encoding**

```python
categorical_columns= [col for col in df.select_dtypes('object').columns]
categorical_columns
```

```python
numerical_columns= [col for col in df.select_dtypes(['int64','float64']).columns]
numerical_columns
```

```python
#One-Hot Encoding of Categorical Column Using Pandas library
categorical_columns= [col for col in df.select_dtypes('object').columns]
one_hot_encoded_data = pd.get_dummies(df,categorical_columns,drop_first= True,dtype= int )
Main_data=one_hot_encoded_data
Main_data.head(5)
```

### **Feature Scaling**

```python
#Separating Feature and Target matrices
X = Main_data.drop(['Attrition'], axis=1)
y=Main_data['Attrition']
```

```python
# import module
from sklearn.preprocessing import MinMaxScaler
# scale features
scaler = MinMaxScaler()
model_X=scaler.fit_transform(X)


# print scaled features
print(model_X)
```

### **Splitting of Data into training & Testing data**

```python
# Split the data into Training set and Testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(model_X,y,test_size =0.2,random_state=42)
```
